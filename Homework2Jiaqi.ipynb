```python
x = np.linspace(-4, 4, 100)   #still need to understand on how to print this step

def phi(r, r0, aB):
    return np.exp(-abs(r-r0)/aB)

phiAVec = []   
phiBVec = []  
phiAB_ABVec = [] 
phiAB_BVec = []   

r0A = 1   # Distance between the atoms is 2 units
aB = 1
r0B = -1

for r in x:
    phiAVec.append(phi(r, r0A, aB))  # Wave function of A-atomic orbital
    phiBVec.append(phi(r, r0B, aB))  #   "         "    B-atomic orbital
    
    phiAB_AB = v[0,0]*phi(r,r0A, aB) + v[1,0]*phi(r,r0B, aB) # LCAO  
    phiAB_B = v[0,1]*phi(r,r0A, aB) + v[1,1]*phi(r,r0B, aB)
    
    phiAB_ABVec.append(phiAB_AB)
    phiAB_BVec.append(phiAB_B)

plt.figure()
plt.plot(x,phiAVec,'r',label='atom A')
plt.plot(x,phiBVec,'b',label='atom B')
plt.axvline(x=0.0, color='k', linestyle='--', linewidth=1.0)
plt.axvline(x=1.0, color='k', linestyle='--', linewidth=1.0)
plt.axvline(x=-1.0, color='k', linestyle='--', linewidth=1.0)
plt.title("Atomic orbital wave functions")
plt.ylabel("E (eV)")
plt.xlabel("x")
plt.legend()

plt.figure()
plt.plot(x,phiAB_ABVec,'r')
plt.plot(x,phiAB_BVec,'b')
plt.axvline(x=0.0, color='k', linestyle='--', linewidth=1.0)
plt.axvline(x=1.0, color='k', linestyle='--', linewidth=1.0)
plt.axvline(x=-1.0, color='k', linestyle='--', linewidth=1.0)
plt.title("Bonding and Anti-bonding eigen vectors")
plt.ylabel("E (eV)")
plt.xlabel("x")


plt.figure()
plt.axhline(w[0],0,1,color="C0",label ='Bonding')
plt.axhline(w[1],0,1,color="C1",label ='Anti-bonding')
plt.ylabel("E (eV)")
plt.title("Bonding and Anti-bonding energy levels ")
plt.legend()
plt.show()
```
