# Band structure of Linear chain of atom.

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

E0= 0  # eV
t=1    # hopping energy
d=1    # unit distance

K=np.linspace(-np.pi, np.pi,500)

Ev=[]
for k in K:
    E = -2*t*np.cos(k*d)
    Ev.append(E)

###############################
# Plot the 1d band structure  #
###############################

plt.plot(K,Ev,'r-',linewidth=5)    
plt.xlabel('Kx',fontsize=20)
plt.ylabel('Energy(eV)',fontsize=20)
plt.axhline(y=0.0, color='k', linestyle='--', linewidth=3.0)
plt.title('Band structure:E(k)',fontsize=20)
plt.show()
```

# Problem.2 : Band structure of a linear chain of atoms using a unit cell with twice its size.

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

E0 = 0   #eV
t = 1    #eV
d = 1    #unit distance

K = np.linspace(-np.pi,np.pi, 1000)
Ev = []

for k in K:
    f=2*np.cos(k*d)
    H=[[E0,t*f],[t*np.conjugate(f),E0]]
    w,v=LA.eig(H)
    Ev.append(w)
    #print(Ev)
#############################
## Plot the band structure  #
#############################
plt.plot(K,Ev,'.-r')
plt.axhline(y=0.0, color='k', linestyle='--', linewidth=3.0)
plt.xlabel('Kx',fontsize=20)
plt.ylabel('Energy(eV)',fontsize=20)
plt.title('Band structure E(k)',fontsize=20)
```
