```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

EA = -1 #eV
EB = 1
t = 1 #eV
d = 1 #unit distance

d1 = d
d2 = -d

k = np.linspace(-np.pi,np.pi, 1000)
Ev = []

for i in range(1000):
    kx = k[i]
    f = 2*t*np.cos(kx*d)#np.exp(1j*kx*d1)+ np.exp(1j*kx*d2)
    H = np.array([[EA,f],[t*np.conjugate(f),EB]])
    #print('Hamiltonian =',H)
    w, v = LA.eig(H)
    #print('Eigen values =',w)
    Ev.append(w)
    
#print(Ev)
plt.plot(k,Ev,'-b')
plt.axhline(y=0.0, color='k', linestyle='--', linewidth=1.0)
plt.xlabel('-----Kx----')
plt.ylabel('-----Energy (eV)----')
plt.title('Band structure :E(k)')
plt.show()
```

```python
#1D complex linear chain with different distance, hopping, onsite energy.
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
EA = 0.1 #eV onsite energy
EB = 0.2 #eV
t1 = 1 #eV hopping
t2 = 1 #eV
a = 1 #unit distance

d1 = a
d2 = -a

k = np.linspace(-np.pi,np.pi, 1000)
Ev = []

for i in range(1000):
    kx = k[i]
    f = t1*np.exp(1j*kx*d1)+ t2*np.exp(1j*kx*d2)
    H = np.array([[EA,t*f],[t*np.conjugate(f),EB]])
    #print('Hamiltonian =',H)
    w, v = LA.eig(H)
    #print('Eigen values =',w)
    Ev.append(w)
    #print(Ev)
    
plt.plot(k,Ev,'-b')
plt.axhline(y=0.0, color='k', linestyle='--', linewidth=1.0)
plt.xlabel('-----Kx----')
plt.ylabel('-----Energy (eV)----')
plt.title('Band structure :E(k)')
#plt.ylim([-0.5 ,0.5])
plt.show()
#Question:Whiy it says the eigenvalues is complex 
#but still give the same result of band structure
```
