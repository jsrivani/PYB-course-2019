$ {\bf Problem \,\, 2.}$ Using brick lattices build the tight-binding Hamiltonian with 
hopping parameter $t=-1~$eV between nearest neighbors for (a) a rectangular flake of
$N \times M$ sizes where $N=5$ columns and $M=4$ rows. (b)Build the Hamiltonian of a
pentacene, (c) hexacene, and (d) heptacene. (e) Produce the Hamiltonian of a phenyl 
group quantum dot consisting of one 7-5-7-5 rings.

# 2a)  Hamiltonian For Brick Lattice.

```python
###############################################################
###############################################################
##     Hamiltonian For Brick Lattice                          #
###############################################################
###############################################################


import numpy as np 
from numpy import linalg as LA
import matplotlib.pyplot as plt

N=5  #Total number of Columns.
M=4  #Total Number of Rows.



print('Total Number of Atom=',N)

t=-1   # hopping energy in eV

print('Hopping energy of neighbour site=',t)

#### Initiation of the Hamiltonian ####
H=np.zeros((N*M,N*M))

###################################
###################################
#### Construct the Hamiltonian ####
###################################
###################################

for i in range (1,N*M):
    if(i/5==int(i/5)):
        H[i,i+1]=0
        H[i+1,i]=0
    else:
        H[i-1,i]=t
        H[i,i-1]=t
        
for i in range(1,N*M):
    
    if(i+5<N*M and (i/2)!=int(i/2)):        
        H[i,i+5]=t
        H[i+5,i]=t
        #print(i)
print('The Hamiltonian:')
print('H=')
print(H)

##############################################
##############################################
######### Energy Spectrum Calculation ########
##############################################
##############################################
w,v =LA.eig(H)  # Solving the Hamiltonian 

#w=Eigen Value 
#v=Eigen Vector

print('Eigen Value of the system:')
print(w)

plt.figure()
for i in range (0,N*M-1):
    plt.axhline(w[i],0,1)
    plt.ylabel('Energy',fontsize=18)
plt.show()
```
