$ {\bf Problem \,\, 2.}$ Using python generate a $10 \times 10$ lattice around (0,0) using the notion of
lattice vectors and represent the lattice sites, (a) for a square lattice, and (b) for a honeycomb lattice, 
and (c) rotate them by 20$^\circ$ degrees.

```python
import numpy as np 
from numpy import linalg as LA
import matplotlib.pyplot as plt

Nx=np.arange(-5,6,1)
Ny=np.arange(-5,6,1)

acc=0.145

######### Square Lattice  ############
a1_sq=np.array([1, 0])  # Lattice Vector 
a2_sq=np.array([0, 1])  # Lattice Vector  

for i in Nx:
    for j in Ny:
        r=i*a1_sq+j*a2_sq
        plt.plot(r[0],r[1],'ro')

plt.xlabel('$N_x$',fontsize=18)
plt.ylabel('$N_y$',fontsize=18)   
plt.title('Square Lattice',fontsize=18)

####### Honeycomb Lattice #########
a=0.246    
a1_hex=np.array([a,0])
a2_hex=np.array([a/2,np.sqrt(3)*a/2])

plt.figure()
for i in Nx:
    for j in Ny:
        r=i*a1_hex+j*a2_hex
        plt.plot(r[0]-acc/2,r[1]-acc/2,'ro')
        plt.plot(r[0]+acc/2,r[1]+acc/2,'ro')
print(a1_hex[0],a1_hex[1])
```
